# Vector quantization
Compress images with K set to 8, 16, 32 and 64.

```python
import kmeans
from matplotlib.pyplot import imread
img = imread('sample1.jpg')
fea = img.reshape((img.shape[0]*img.shape[1], 3)).astype(np.float64)
# YOUR (SEVERAL LINES) CODE HERE
# begin answer
print("原图片：")
plt.imshow(fea.reshape(img.shape).astype(np.uint8))
plt.show()
print("进行K聚类以后：")
print(img.shape)
# 利用Kmeans对图片进行压缩的基本思路是：找到K个中心点
# 用中心点的颜色来代表该类里面所有的颜色，这样就减小了图片的存储空间
for k in [1, 2, 4, 8]:
    print("k = " + str(k) + "时候的压缩情况")
    idx, cters, _ = kmeans.kmeans(fea.copy(), k)
    new_img = cters[idx]
    plt.imshow(new_img.reshape(img.shape).astype(np.uint8))
    plt.show()
# end answer


```
